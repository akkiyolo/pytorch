### Tensor Device Creation in PyTorch
#### Device in PyTorch
A device is the physical hardware location where a tensor is stored and processed. PyTorch supports two main device types:
"cpu" — Your computer's central processor
"cuda" — NVIDIA GPU (Graphics Processing Unit)
By default, PyTorch creates tensors on the CPU, unless specified otherwise.

* Training deep learning models on the CPU is very slow, especially for large datasets or big models.
* Using a GPU can accelerate operations 10x to 100x.
* You must move both the model and data to the same device (usually GPU) to avoid runtime errors.

```python
import torch
```

```python
x=torch.tensor([1.0,2.0])
x,x.device
```

```python
x=torch.tensor([1.0,2.0],device="cpu")
x.device
```

```python
device=torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")
```

```python
x=x.to(device)
x.device
```
