Backpropagation = the math trick that automatically figures out exactly how much each weight in the network caused the final error, then tells the optimizer how to fix them.

The chain rule is the magic math shortcut that lets backpropagation “travel backwards” through all the layers by multiplying tiny derivatives together.

```python
x=torch.tensor(1.0)
y=torch.tensor(2.0)

w=torch.tensor(1.0,requires_grad=True)

# forward pass and compute the loss

y_hat=w*x
loss=(y_hat-y)**2

print(loss)
```

```python
# backward pass

loss.backward()
print(w.grad)
```
